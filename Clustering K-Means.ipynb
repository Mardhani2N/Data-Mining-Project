1. Algoritma *Clustering* K-Means

*Clustering* adalah proses pengelompokan sekumpulan objek fisik ataupun abstrak
ke dalam kelas-kelas yang mempunyai kemiripan (Han dan Kamber, 2001). Terdapat
berbagai macam jenis *clustering*, salah satunya *Clustering* K-Means. K-Means
merupakan salah satu metode data *clustering* nonhierarki yang berusaha mempartisi
data yang memiliki karakteristik sama di kelompokkan ke dalam satu cluster yang
sama.

Menurut Fauzi (2017), dalam melakukan *clustering* dengan algoritma K-Means
dibagi menjadi beberapa tahapan sebagai berikut. <br>
a. Tentukan jumlah *cluster* yang ingin dibentuk dari sebuah dataset. <br>
b. Tentukan lokasi tengah atau *cluster* *centroid* secara acak. <br>
c. Untuk setiap baris data, temukan *cluster* yang paling dekat dengan pusat *cluster*. <br>
d. Untuk setiap *cluster* k, temukan *cluster* *centroid* dan diperbaharui lokasi masing-masing pusat *cluster* untuk mendapatkan nilai *centroid* yang baru. <br>
e. Ulang langkah ke-3 sampai langkah ke-5 sampai ditemui *cluster* yang konvergen sebagai akhir dari *clustering*.

Setelah itu dilakukan scalling pada data dengan menggunakan MinMaxScaler. Proses ini juga dapat dikatakan sebagai proses menormalisasikan fitur/variabel, yang dalam hal ini adalah atributnya.
```python
# melakukan scaling data sekaligus menghapus data pada kolom nama_daerah
minmax = preprocessing.MinMaxScaler().fit_transform(data_keamanan1.drop('nama_daerah', axis = 1))
```

Setelahnya data yang sudah di-scalling akan dibuat ke dalam bentuk dataframe baru.

```python
# menampilkan data hasil scaling ke dalam bentuk dataframe
data_keamanan2 = pd.DataFrame(minmax, index = data_keamanan1.index, columns = data_keamanan1.columns[1:])
data_keamanan2
```

<hr>

Sebelum melakukan *clustering*, maka perlu menentukan jumlah cluster terlebih dahulu.

## Penentuan Jumlah Cluster

Dalam penyelesaian studi kasus ini digunakan Metode Elbow sebagai metode untuk
menentukan jumlah klaster terbaik atau paling optimal berdasarkan hasil perbandingan antara jumlah klaster yang akan membentuk siku pada suatu titik.

```python
# menghilangkan pesan peringatan (warning)
import warnings
warnings.filterwarnings('ignore')
```

```python
# menentukan nilai cluster dengan metode elbow
wcss = []
for i in range(1, 10):
  kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state = 42).fit(data_keamanan2)
  wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Within-Cluster Sum of Square')
```

Berdasarkan grafik yang terlihat diperoleh bahwa pada titik ke-4 merupakan titik di mana elbow atau siku terbentuk ketika grafik mulai melandai, sehingga jumlah klaster yang dipilih adalah empat.

<hr>

Kemudian dilakukan *clustering* dengan K-Means

## Clustering

```python
# melakukan clustering dengan jumlah cluster yaitu 4
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data_keamanan2)
```

```python
# menambahkan kolom cluster (hasil clustering) ke dalam dataframe
data_keamanan2['cluster'] = kmeans.labels_
data_keamanan2
```

*clustering* K-Means dapat digunakan untuk mengelompokkan tingkat keamanan dari 34 provinsi di Indonesia pada tahun 2018 berdasarkan jumlah kejahatan menurut kepolisian daerah pada tahun 2018, risiko penduduk terkena kejahatan per 100.000 penduduk menurut kepolisian daerah pada tahun 2018, jumlah desa/kelurahan menurut jenis upaya warganya untuk menjaga keamanan menurut provinsi pada tahun 2018, dan jumlah desa/kelurahan menurut keberadaan pos polisi dan kemudahan akses ke pos polisi menurut provinsi pada tahun 2018.

Terdapat 6 provinsi yang masuk ke dalam *cluster* pertama yaitu Gorontalo, Papua, Papua Barat, Sulawesi Tengah, Sulawesi utara, dan Sulawesi Barat.
Kemudian sejumlah 22 provinsi termasuk ke dalam *cluster* kedua yaitu Aceh, Riau, Jambi, Sumatera Selatan, Bengkulu, Lampung, Kepulauan Bangka Belitung, Kepulauan Riau, Daerah Istimewa Yogyakarta, Banten, Bali, Nusa Tenggara Barat, Nusa Tenggara Timur, Kalimantan Barat, Kalimantan Tengah, Kalimantan Selatan, Kalimantan Timur, Kalimantan Utara, Sulawesi Tenggara, Sulawesi Barat, Maluku, dan Maluku Utara. Terdapat 3 provinsi di *cluster* ketiga yaitu Jawa Barat, Jawa Tengah, dan Jawa Timur. Sehingga, sisanya 3 provinsi masuk ke dalam *cluster* keempat yaitu DKI Jakarta, Sulawesi Selatan, dan Sumatera Utara.
